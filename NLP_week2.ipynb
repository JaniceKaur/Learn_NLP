```python
 # Filtering Text- T o remove morphological variants or missplelled Words in text
def unusual_words(text):
        text_vocab=set(w for w in text if w.isalpha())
        english_vocab=set(w for w in words.words())
        unusual_words=text_vocab.difference(english_vocab)
        return unusual_words
        
```

```python
unusual_words(gutenberg.words('austen-emma.txt'))
```

```python
freq_letters=nltk.FreqDist('egivrvonl')
wordlist=words.words()
obligatory='r'
[w for w in wordlist if len(w)>=6 and obligatory in w and nltk.FreqDist(w)<=freq_letters]
```

```python
# FIRE - Forum for Information Retrieval and Evaluation
stop=stopwords.words(fileids='english')
def fraction_not_stopwords(text):
    content_not=[w for w in text if w not in stop]
    return len(content_not)/len(text)
```

```python
fraction_not_stopwords(gutenberg.words(fileids='austen-emma.txt'))
```

```python
# condn frequency distributions - divides into groups then finds the frequency of those groups
cfd=nltk.ConditionalFreqDist((fileid,name[-1])
                        for fileid in names.fileids()
                            for name in names.words(fileid))
```

```python
cfd.plot()
```
